## Let's plot something

```python
x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)
plt.plot(x, y, color='#e433ff')
```

## Plotting Styles

Line Plots

<code> '-'             solid line style
 '--'            dashed line style
 '-.'            dash-dot line style
 ':'             dotted line style
 'steps' | 'steps-pre' | 'steps-mid' | 'steps-post'
 </code>
 
 

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(14, 14))
# plt.plot?
ax[0, 0].plot(x, y)
ax[0, 1].plot(x, y, '--')
ax[0, 2].plot(x, y, linestyle=':')
ax[1, 0].plot(x, y, linestyle=':', drawstyle ='steps')
ax[1, 1].plot(x, y, linestyle=':', drawstyle ='steps-pre')
ax[1, 2].plot(x, y, linestyle=':', drawstyle ='steps-pre')
ax[1, 2].plot(x, y, linestyle=':', drawstyle ='steps-mid')
ax[1, 2].plot(x, y, linestyle=':', drawstyle ='steps-post')
```

### Error bars
Can plot the error of each individual point in both x and y direction

```python
xerr = np.sqrt(x)
yerr = 0.1

fig, ax = plt.subplots(ncols=3, figsize=(16, 4))

ax[0].errorbar(x, y, xerr=xerr, yerr=yerr)
# no error caps, color the error bars differently than the data:
ax[1].errorbar(x, y, xerr=xerr, yerr=yerr, capsize=0, ecolor='gray')
# don't plot values? Just error bars? set fmt='none'
ax[2].errorbar(x, y, xerr=xerr, yerr=yerr, capsize=0, ecolor='gray', fmt='none')
```

### Histogram
Bins the data and plots their counts as a density

```python
#Example: 
x = np.random.normal(size=10000)
fig, ax = plt.subplots()
h, bins = np.histogram(x, bins=16)
ax.plot(bins[1:], h,  linestyle='-', drawstyle ='steps-pre')
plt.draw()
```

## Plotting with Data

### Let's read in a text file (should already in your jupyter notebook directory


```python
#Here is a function that can read in a file.
#Let's look at it line by line
def readfile(filename):
    #create our return variables
    wavelength = []
    flux = []
    
    #open up the file
    inputfile = open(filename, 'r')
    
    #iterate over each row in the file
    for row in inputfile:
        #lets skip the header
        if '#' not in row:
            
            #split the row based off spaces. If csv you can split(',')
            splitrow = row.split()
            #the wavelength are the first item.. convert to a float
            wavelength.append(float(splitrow[0]))
            #the flux value is the second item.. convert to a float
            flux.append(float(splitrow[1]))
            
    #return our values!
    return wavelength, flux
```

**Exercise:** <font color=blue>Let's use this function to read in the file: 'solar_spectrum_3.txt' and plot it using matplotlib.pyplot package!</font>

* set the xlim to 400-2500 ... plt.xlim(low, high)
* set the ylim to -0.1 to 1.5 ... plt.ylim(low, high)
* and plot it!

```python
waves, flux = readfile('solar_spectrum_3.txt')
plt.plot(waves, flux, color='red')
plt.xlabel('Wavelength')
```
